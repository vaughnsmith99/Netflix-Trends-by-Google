```python
import pandas as pd
import matplotlib as plt
```

```python
netflix_df= pd.read_csv('netflix_titles.csv').dropna()
trends_df = pd.read_csv('trends.csv').dropna()
```

```python
netflix_df
```

```python
trends_df
```

```python
merged = pd.merge(left=trends_df, right=netflix_df, how='inner', left_on=['query', 'year'], right_on=['title', 'release_year'])
merged
```

```python
merged.loc[(merged['category'] == 'Trending Movies & TV Shows') & (merged['location'] == 'Global')]
```

```python

merged.loc[merged['category'] == 'Trending Movies & TV Shows']
```

```python
filtered_merged = merged.loc[(merged['category'] == 'Trending Movies & TV Shows') |
           (merged['category'] == 'Popular Movies')|
           (merged['category'] == 'Movies') |
           (merged['category'] == 'Movies/TV Series')|
           (merged['category'] == 'Top Trending Movies') |
           (merged['category'] == 'Trending Movies')]
```

```python
filtered_merged
```

```python
cleaned_filtered_merged = filtered_merged[["title", "location","year","country","duration","rating","listed_in","rank"]].dropna()
cleaned_filtered_merged = cleaned_filtered_merged.rename(columns={'location': 'country trending'})
cleaned_filtered_merged = cleaned_filtered_merged.rename(columns={'listed_in': 'genres'})
cleaned_filtered_merged = cleaned_filtered_merged.rename(columns={'rank': 'google search rank'})
cleaned_filtered_merged = cleaned_filtered_merged.rename(columns={'country': 'origin'})
cleaned_filtered_merged


```

```python
cleaned_filtered_merged.to_csv('cleaned.csv', index=False)
```
